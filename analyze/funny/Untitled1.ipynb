```python

def split_dataframe(df1, df2, chunk1=6000, chunk2=9000):
    chunks1 = []
    chunks2 = []
    chunks = []
    num_chunk1 = len(df1) // chunk1 + 1
    num_chunk2 = len(df2) // chunk2 + 1
    for i in range(num_chunk1):
        chunks1.append(df1[i*chunk1:(i+1)*chunk1])
    for i in range(num_chunk2):
        chunks2.append(df2[i*chunk2:(i+1)*chunk2])
        
    # print(len(chunks1))
    # print(len(chunks2))
    chunks = [pd.concat([element[0], element[1]]) for element in zip(chunks1, chunks2)]
    remain_chunks = chunks2[len(chunks1) + 1:]
    remain_df = pd.concat(remain_chunks)
    num_remain_chunk = len(remain_df) // 15000 + 1
    for i in range(num_remain_chunk):
        chunks.append(remain_df[i*15000:(i+1)*15000])
        
    return chunks
        
    
```

```python
def merge(df1, df2):
    df = [df1, df2]
    n = len(df)
    nrows = df[0].shape[0]
    ncols = df[0].shape[0]
    A = pd.concat(df).values.reshape(nrows, -1, ncols)
    sidx = np.random.rand(ncols, n).argsort(1)
    output_arr = A[np.arange(nrows)[:, None], sidx, :]
    return pd.DataFrame(output_arr)


data = merge(snappy_not_sent, njv_correct)

data
```
