```python
import pandas as pd
import numpy as np
import os
import openpyxl

# print(os.getcwd())
os.chdir('./snp')
# print(os.getcwd())


# read very large excel file 
wb = openpyxl.load_workbook('data-tham1.xlsx', read_only=True)
# print(wb)
rows = wb.active.rows
first_row = [cell.value for cell in next(rows)]

dt = []
for row in rows:
    record = {}
    for key, cell in zip(first_row, row):
        record[key] = cell.value
    dt.append(record)

# data = pd.read_excel('data-tham.xlsx', index_cols=0, sheet_name='Sheet1', engine="openpyxl")
data = pd.DataFrame(dt)
data.head()
```

```python
def validate_phone(s: str) -> str:
    if '84' in s:
        return s.replace('84', '0')
    return s


data['To Contact'] = data['To Contact'].astype(str).apply(lambda x: validate_phone(x))
data.head()
```

```python
def get_isp_name(maps, key):
     try:
        return maps.get(int(key[1:3]), np.nan)
     except ValueError:
        return np.nan
    
data['ISP'] = data['To Contact'].astype(str).apply(lambda x: get_isp_name(isp_maps, x))

data.head()
```

```python
# data.to_excel('data-tham1.xlsx')

def split_df(df, chunksize=10000):
    chunks = []
    num_chunks = len(df) // chunksize + 1
    for i in range(num_chunks):
        chunks.append(df[i*chunksize:(i+1)*chunksize])
    return chunks

# chunksize = 10000
# chunks = len(data) // chunksize
chunks = split_df(data)

i = 0
for chunk in chunks:
    chunk.to_excel('data-tham_{:02d}.xlsx'.format(i), index=False)
    i += 1

```
