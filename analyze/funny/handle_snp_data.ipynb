```python
# cleaning data of nguoi_nhan
nguoi_nhan['Số điện thoại'] = nguoi_nhan['Số điện thoại'].astype(str).apply(lambda x: '0' + x)
nguoi_nhan['Số điện thoại']
```

```python
# cleaning data of nguoi_gui
def validate_phone(s: str) -> str:
    if '+84' in s:
        return s.replace('+84', '0')
    return s
    
    
nguoi_gui['SĐT gửi hàng'] = nguoi_gui['SĐT gửi hàng'].apply(lambda x: validate_phone(x))
nguoi_gui.head()
```

```python
nguoi_gui.drop_duplicates(subset=['SĐT gửi hàng'], keep='first', inplace=True)
nguoi_gui.reset_index(inplace=True, drop=True)
nguoi_gui.head()
```

```python
nguoi_nhan.drop_duplicates(subset=['Số điện thoại'], keep='first', inplace=True)
nguoi_nhan.reset_index(inplace=True, drop=True)
nguoi_nhan.head()
```

```python
def get_isp_name(maps, string):
    # get isp name from mapping dict
    try:
        return maps.get(int(string[1:3]), np.nan)
    except ValueError:
        return np.nan
    
nguoi_gui['ISP'] = nguoi_gui['SĐT gửi hàng'].astype(str).apply(lambda x: get_isp_name(isp_maps, x))
nguoi_gui.head()
```

```python
nguoi_nhan['ISP'] = nguoi_nhan['Số điện thoại'].astype(str).apply(lambda x: get_isp_name(isp_maps, x))
nguoi_nhan.head()
```

```python
# split large file into chunks file
def split_df(df, chunksize=10000):
    chunks = []
    num_chunks = len(df) // chunksize + 1
    for i in range(num_chunks):
        chunks.append(df[i*chunksize:(i+1)*chunksize])
    return chunks

# chunksize = 10000
# chunks = len(data) // chunksize
chunks = split_df(data_nguoi_nhan)

i = 0
for chunk in chunks:
    chunk.to_excel('data-nguoi-nhan_{:02d}.xlsx'.format(i), index=False)
    i += 1
```
