```python
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt


os.chdir('/Volumes/UNTITLED/')

list_cols = ['NVPTTT', 'NVXL', 'TrangThai', 'BuuCuc', 'GoiCuoc', 'Loai', 'Tinh', 'Quan', 'TGCapNhat', 'MaDH', 'TrongLuong', 'TenShop', 'MaKho', 'KichThuoc', 'Null1', 'MaCD', 'Null2', 'LoaiHang', 'Null3']

extension = 'csv'
all_filenames = [i for i in glob.glob('2007*.{}'.format(extension))]
print(all_filenames)

# csv_file = pd.read_csv('200601_Hồ Chí Minh.csv', names=list_cols)
# csv_file_dropped = csv_file.loc[~(csv_file['BuuCuc'] != 'Bưu Cục 37 Bế Văn Cấm-Q.7-HCM 01')]
# csv_file_dropped

# read and concatenate all csv files
combine_csv = pd.concat([pd.read_csv(f, names=list_cols) for f in all_filenames])

# filter by BuuCuc col
combine_csv_dropped = combine_csv.loc[~(combine_csv['BuuCuc'] != 'Bưu Cục 37 Bế Văn Cấm-Q.7-HCM 01')]
# drop all names that's not the first names
csv_result = combine_csv_dropped.loc[(combine_csv_dropped['NVPTTT'] != 'NVPTTT')]
csv_result.head(10)
```
